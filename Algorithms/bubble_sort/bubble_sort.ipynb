# Bubble sort 
this algorithm will loop a conditioinal whether or not there was a swap, once there is no swaps then the loop will stop 
    - inside the loop we will loop again through all the values  up to but not including the last values added, so it more efficient since we know that
    if the current value is greater than the previous value, then a swap should occurr and swap is changed to True

    * Its important to know that this algorithm has a runtime of O(n^2) therefore is really slow

```python
def bubble_sort(my_list: list) -> list:
    '''Bubble sort algorithm takes a list and it sorts it out'''
    new_list = my_list.copy()
    swap = True 
    while swap: 
        swap = False
        end = 0
        for i in range(1, len(my_list) - end ):
            if my_list[i -1] > my_list[i]:
                my_list[i -1], my_list[i] = my_list[i], my_list[i -1]
                swap = True
        end += 1
    return my_list
```

```python
#test algorithm

array = [4,1,43,11,78,9,0,11,12,24]
#expected out put after array should be [0,1,4,9,11,11,,12,24,43,78]
assert(bubble_sort(array) == sorted(array))
print(bubble_sort(array))
```
