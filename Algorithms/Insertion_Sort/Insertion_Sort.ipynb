# Insertion sort
* Insertion sort has a big O(n^2), important to know that in small cases insertion sort is really fast but as soon as the import grows so does its time complexity

insertion sort starts at index 1, it checks the previous values, up to the beginning, if the value is greater then it swaps the values until no greater then swap to value selected


```python
def insertion_sort(my_list: list) -> list:
    new_list = my_list.copy() #just so original copy is not modified because of mutability

    for i in range(1, len(new_list)): # loop through index starting at 1

        j = i
        while j -1 >= 0 and new_list[j -1] > new_list[j] : # look in reverse until element at j -1 is less than element at j[i] or j -1 is a negative value
            
            new_list[j - 1], new_list[j] = new_list[j], new_list[j -1]
            j -= 1
    return new_list
```

```python
#unit testing
from random import randrange

array = [randrange(1_000_000) for i in range(100_000)]
sorted_array = sorted(array)
insertion_sort_array = insertion_sort(array)

assert (insertion_sort_array == sorted_array)
print(insertion_sort_array)
```
